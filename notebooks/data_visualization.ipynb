# Data Visualization
Exploring and checking the data, visualization, 2d and 3d renders.

```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

filePath = Path('../data/processed/ST0202_full_cube_notnormal.npy')

array = np.load(filePath)

print(array.dtype,array.shape)
print(array.size)
```

```python
for x in range (10,200,30):
    sliced = array[:,:,x]
    plt.imshow(sliced,cmap='seismic',aspect='auto')
    plt.colorbar()
    plt.title(f'Inline Sliced at X ={x}')
    plt.xlabel("Crossline Index")
    plt.ylabel("Depth Index")
    plt.show()
```
